```python
import numpy as np
from Snake.Game import Game
import neat
import pickle
import copy
import math
```

```python
game = Game(30, 30)
```

```python
def dir_from_float(flt):
    if flt < -.3:
        return 2
    if flt > 0.3:
        return 1
    return 0
```

```python
def eval_genomes(genomes, config):
    for genome_id, genome in genomes:
        net = neat.nn.RecurrentNetwork.create(genome, config)
        GAME = Game(30, 30)
        genome.fitness = 0

        while GAME.snake:
            prediction = net.activate(GAME.sniff())[0]
            direction = dir_from_float(prediction)
            GAME.step(direction)
        genome.fitness = 2**GAME.eaten * GAME.steps 
```

```python
def run(conf):
    config = neat.Config(neat.DefaultGenome, neat.DefaultReproduction,
                         neat.DefaultSpeciesSet, neat.DefaultStagnation,
                         conf)

    p = neat.Population(config)

    p.add_reporter(neat.StdOutReporter(True))
    stats = neat.StatisticsReporter()
    p.add_reporter(stats)
    p.add_reporter(neat.Checkpointer(5))

    winner = p.run(eval_genomes, 300)

    # Display the winning genome.
    print('\nBest genome:\n{!s}'.format(winner))

    return winner
```

```python
winner = run("config")
```

```python
cmap = colors.ListedColormap(['white', 'red', 'green', '#cb7341'])
boundaries = [-.9, -0.75, 0.5, .9]
norm = colors.BoundaryNorm(boundaries, cmap.N, clip=True)
```

```python
sg = copy.deepcopy(game)
```

```python
sg.reset()
```

```python
input_size = 12
```

```python
def update_ai(ax):
    def f(*args, **kwargs):
        if not sg.snake:
            return
        a = winner.activate(sg.sniff())[0]
        sg.step(dir_from_float(a))
        ax.cla()
        ax.imshow(sg.map, cmap=cmap, interpolation='nearest', animated=True, extent=(0,sg.h,0,sg.w))
        ax.set_title(f"AISnake: {sg.status}, steps: {sg.steps}, lives: {sg.score}, apples: {sg.eaten}\n bad steps: {sg.bad_steps} fitness: {int(math.sqrt(sg.fitness))}")
        ax.axis('off');
        return ax
    return f        
```

```python
fig, ax = plt.subplots();
f = update_ai(ax);
anim = animation.FuncAnimation(fig, f, interval=175, blit=False);
plt.axis('off');
plt.show()
```
